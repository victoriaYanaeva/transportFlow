```python
import numpy as np
import scipy as scp

import csv
import scipy.io

u0 = np.zeros(3)
alpha = 100     
#print(alpha)
beta = 0.001
w = 0.01
u0[0], u0[1], u0[2] = alpha, beta, w

N = []
L = []
W = []

def attraction_function(C):
    f = np.exp(-alpha*C**beta)
    return f
def corresp_matrix(A,B,L,W,C):
    x = np.zeros((24,24))
    for i in range(len(L)):
        for j in range(len(W)):
            x[i,j] = A[i]*B[j]*L[i]*W[j]*C[i,j]**w*attraction_function(C[i,j])
    return x
def count_sum(x,k):
    sum_x = np.sum(x,axis=k)
    return sum_x

FILENAME = "table.csv"
Number, L, W = [], [], []
with open(FILENAME, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        #print(row.split(';'))
        k = 0
        row[k] = row[k].split(';')
        for n in range(24):
            Number.append(float(row[k][3*n]))
            L.append(float(row[k][3*n+1]))
            W.append(float(row[k][3*n+2]))
Number, L, W =np.array(Number), np.array(L), np.array(W)

size = len(L)

FILENAME = "AvTimeAll.csv"
C = np.zeros((size,size))
with open(FILENAME, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        a = row[0]
        b = a.split(':')
        b0 = b[0].split(' ')
        b.pop(0)
        C_row = b0+b
        C[int(C_row[0])-1, int(C_row[1])-1] = float(C_row[2])

F1, F2, F3 = 0, 0, 0
J = np.zeros((3,3))
F = np.zeros(3)
f  = attraction_function(C)
A0 = np.zeros(size)
B0 = np.zeros(size)
A = np.zeros(size)
B = np.zeros(size)
x0 = np.zeros((size,size))

for i in range(size):
    A0[i] = 1./L[i]
for j in range(size):
    B0[j] = 1./W[j]
#print(A0[0])
summa_A = 0.
summa_B = 0.
step = 0
norm0, norm = 0, 0
x0 = corresp_matrix(A0,B0,L,W,C)
print(x0[0,0])

for j in range(size):
    summa_B = 0.
    for i in range(size):
        summa_B += A0[i]*L[i]*C[i,j]**w*f[i,j]
    B[j] = 1./summa_B

B0 = B.copy()
for i in range(size):
    summa_A = 0.
    for j in range(size):
        summa_A += B0[j]*W[j]*C[i,j]**w*f[i,j] 
    A[i] = 1./summa_A
x = corresp_matrix(A,B,L,W,C)
print(x[0,0])


norm_w = count_sum(x,0) - W
norm_l = count_sum(x,1) - L
norm_w0 = count_sum(x0,0) - W
norm_l0 = count_sum(x0,1) - L
print(np.linalg.norm(norm_w))
while(np.linalg.norm(norm_w,2)>1e-6):
    #print(np.linalg.norm(norm_w,2),step)
    A0 = A.copy()
    
    for j in range(size):
        summa_B = 0.
        for i in range(size):
            summa_B += A0[i]*L[i]*C[i,j]**w*f[i,j]
        B[j] = 1./summa_B
    B0 = B.copy()
    #print(B)
    for i in range(size):
        summa_A = 0.
        for j in range(size):
            summa_A += B0[j]*W[j]*C[i,j]**w*f[i,j] 
        A[i] = 1./summa_A
    x = corresp_matrix(A,B,L,W,C)
    norm_w = count_sum(x,0) - W
    norm_l = count_sum(x,1) - L
    
    step += 1
x = corresp_matrix(A,B,L,W,C)
#print(np.linalg.norm(norm_w))
for i in range(size):
    for j in range(size):
        F1 += -(x[i,j]-Corresp_matrix_etalon[i,j])*x[i,j]*C[i,j]**beta
        dx_dalpha = -x[i,j]*C[i,j]**beta
        F2 += -alpha*(x[i,j]-Corresp_matrix_etalon[i,j])*x[i,j]*C[i,j]**beta*np.log(C[i,j])
        dx_dbeta = -alpha*x[i,j]*C[i,j]**beta*np.log(C[i,j])
        F3 += (x[i,j]-Corresp_matrix_etalon[i,j])*x[i,j]*np.log(C[i,j])
        dx_dw = x[i,j]*np.log(C[i,j])
        delta = x[i,j]-Corresp_matrix_etalon[i,j]
        J[0,0] += (dx_dalpha)**2 + delta*x[i,j]*C[i,j]**(2*beta)
        J[0,1] += dx_dalpha*dx_dbeta + delta*(-dx_dbeta*C[i,j]**beta - x[i,j]*C[i,j]**beta*np.log(C[i,j]))
        J[0,2] += dx_dalpha*dx_dw + delta*(-dx_dw*C[i,j]**beta)
        J[1,1] += (dx_dbeta)**2 + delta*((1./x[i,j])*dx_dbeta**2+dx_dbeta*np.log(C[i,j]))
        J[1,2] += dx_dbeta*dx_dw + delta*dx_dbeta*np.log(C[i,j])
        J[2,2] += dx_dw**2 + delta*dx_dw*np.log(C[i,j])
J[1,0] = J[0,1]
J[2,0] = J[0,2]
J[2,1] = J[1,2]
         
F = np.array([F1, F2, F3])
u = np.zeros(3)
u = np.linalg.solve(J,-F)

#print(u, np.linalg.norm(norm_w), np.linalg.norm(u))
t = np.zeros((size, size))
error = 0
for i in range(size):
    for j in range(size):
        error += (x[i,j]-Corresp_matrix_etalon[i,j])**2
        if Corresp_matrix_etalon[i,j]!=0:
            t[i,j] = (x[i,j]-Corresp_matrix_etalon[i,j])*100/Corresp_matrix_etalon[i,j]
        else:
            t[i,j] = 100
check = error
print(np.max(np.around(np.abs(t),2)), np.min(np.around(np.abs(t),2)), np.max(Corresp_matrix_etalon), np.max(x))
print(check)
np.disp(x)
```
